```python
from matplotlib import pyplot as plt
from scipy import stats
import numpy as np
import pandas as pd
import random
```

```python
ecomm_df = pd.read_excel('e-commerce-q20.xlsx',index_col=None)
ecomm_df.to_csv('ecomm_df.csv', encoding='utf-8')

csvfile = "ecomm_df.csv"
ecomm_csv = pd.read_csv(csvfile)
ecomm_csv
```

```python
ecomm_total = ecomm_csv[["Quarter","Retail Sales(millions of dollars) Total",
                         "Retail Sales(millions of dollars) E-commerce"]]
ecomm_index = ecomm_total.set_index("Quarter")
ecomm_index
```

```python
ecomm_index.plot(kind="bar", figsize=(20,3))
plt.title("Retail Sales Total Vs. E-Commerce")
plt.xlabel("Quarter")
plt.ylabel("Sales")
plt.show()
plt.tight_layout()
```

```python
Quarter_2020 = ecomm_index.loc["2nd quarter 2020(p)"]
quarter = Quarter_2020.keys()

# Create a pie chart based upon the trip duration of that single bike
Quarter_2020.plot(kind="pie", y=quarter, autopct="%1.1f%%",title= "Retail Total Vs. E-comm")
plt.show()
plt.tight_layout()
plt.axis("equal")
```

```python
Quarter_2019 = ecomm_index.loc["2nd quarter 2019(r)"]
quarter = Quarter_2019.keys()

# Create a pie chart based upon the trip duration of that single bike
Quarter_2019.plot(kind="pie", y=quarter, autopct="%1.1f%%",title= "Retail Total Vs. E-comm")
plt.show()
plt.tight_layout()
plt.axis("equal")
```
